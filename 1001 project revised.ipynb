```python
# 1208
# cross_validation
# decision_tree test with max-depth 
# NB test with alpha
# svm test with c
# LR test with C=c, penalty='l{}'.format(norm)
# knn test with n-neigh k 
%matplotlib inline

def xValSVM(dataset, label_name, k, cs):
    auc = {}
    cv = KFold(n=dataset.shape[0], n_folds = k)
    for train_index, val_index in cv:
        dataset_train = dataset.iloc[train_index]
        dataset_validate = dataset.iloc[val_index]
        for c in cs:
            my_svm = svm.SVC(C=c,probability=True)
            my_svm.fit(dataset_train.drop(label_name, 1).values, dataset_train[label_name].values)
            prediction_svm = my_svm.predict_proba(dataset_validate.drop(label_name, 1).values)[:,1]
            fpr_svm, tpr_svm, thresholds_svm = metrics.roc_curve(dataset_validate[label_name].values, prediction_svm)
            auc_c_k = metrics.auc(fpr_svm,tpr_svm)  
            if (auc.has_key(c)):
                auc[c].append(auc_c_k)
            else:
                auc[c] = [auc_c_k]                
    return auc


def xValLR(X_tr, Y_tr, k, cs):

    cv = KFold(n=X_tr.shape[0], n_folds = k)
    aucs = {}

    for train_index, test_index in cv:
        X_tr_f = X_tr.iloc[train_index]
        X_va_f = X_tr.iloc[test_index]
        Y_tr_f = Y_tr.iloc[train_index]
        Y_va_f = Y_tr.iloc[test_index]
        
        for c in cs:
            for norm in [1,2]:
                lr = LogisticRegression(C=c, penalty='l{}'.format(norm))
                lr.fit(X_tr_f,Y_tr_f)
                met = metrics.roc_auc_score(Y_va_f, lr.predict_proba(X_va_f)[:,1])

                if (aucs.has_key((c, norm))):
                    aucs[(c, norm)].append(met)
                else:
                    aucs[(c, norm)] = [met]    
    return aucs


def xValKNN(dataset, label_name, k_fold, c):
    auc = {}
    cv = KFold(n=dataset.shape[0], n_folds = k_fold)
    
    for train_index, val_index in cv:
        dataset_train = dataset.iloc[train_index]
        dataset_validate = dataset.iloc[val_index]
        for i in c:
            my_knn = KNeighborsClassifier(n_neighbors=i)
            my_knn.fit(dataset_train.drop(label_name, 1).values, dataset_train[label_name].values)
            prediction_knn = my_knn.predict_proba(dataset_validate.drop(label_name, 1).values)[:,1]
            fpr_knn, tpr_knn, thresholds_knn = metrics.roc_curve(dataset_validate[label_name].values, prediction_knn)
            auc_c_k = metrics.auc(fpr_knn,tpr_knn)   
            if (auc.has_key(i)):
                auc[i].append(auc_c_k)
            else:
                auc[i] = [auc_c_k]                
    return auc


def xValDT(dataset, label_name, k_fold, max_depth):
    auc = {}
    cv = KFold(n=dataset.shape[0], n_folds = k)
    for train_index, val_index in cv:
        dataset_train = dataset.iloc[train_index]
        dataset_validate = dataset.iloc[val_index]
        for c in max_depth:
            my_DT = DecisionTreeClassifier(max_depth = c)
            my_DT.fit(dataset_train.drop(label_name, 1).values, dataset_train[label_name].values)
            prediction_DT = my_DT.predict_proba(dataset_validate.drop(label_name, 1).values)[:,1]
            fpr_DT, tpr_DT, thresholds_DT = metrics.roc_curve(dataset_validate[label_name].values, prediction_DT)
            auc_c_k = metrics.auc(fpr_DT,tpr_DT)  
            if (auc.has_key(c)):
                auc[c].append(auc_c_k)
            else:
                auc[c] = [auc_c_k] 
    return auc


def xValNB(dataset, label_name, k_fold, alpha):
    auc = {}
    cv = KFold(n=dataset.shape[0], n_folds = k)
    
    for train_index, val_index in cv:
        dataset_train = dataset.iloc[train_index]
        dataset_validate = dataset.iloc[val_index]
        for c in alpha:
            my_NB = BernoulliNB(alpha=c)
            my_NB.fit(dataset_train.drop(label_name, 1).values, dataset_train[label_name].values)
            prediction_NB = my_NB.predict_proba(dataset_validate.drop(label_name, 1).values)[:,1]
            fpr_NB, tpr_NB, thresholds_NB = metrics.roc_curve(dataset_validate[label_name].values, prediction_NB)
            auc_c_k = metrics.auc(fpr_NB,tpr_NB)  
            if (auc.has_key(c)):
                auc[c].append(auc_c_k)
            else:
                auc[c] = [auc_c_k]
    return auc
```

```python
# Decision Tree Classifier

from scipy.stats import *
xval_dict_dt = {'e':[], 'mean':[], 'stdErr':[],'plus':[],'minus':[]}
for i in depth:
    xval_dict_dt['e'].append(i)
    xval_dict_dt['mean'].append(np.array(auc_dt[i]).mean())
    xval_dict_dt['stdErr'].append(sem(np.array(auc_dt[i])))
    xval_dict_dt['plus'].append(np.array(auc_dt[i]).mean()+sem(np.array(auc_dt[i])))
    xval_dict_dt['minus'].append(np.array(auc_dt[i]).mean()-sem(np.array(auc_dt[i])))
res_dt = pd.DataFrame(xval_dict_dt)

fig=plt.figure()
ax=fig.add_subplot(111)

plt.plot(res_dt['e'], res_dt['mean'],label='mean AUC')
plt.plot(res_dt['e'], res_dt['plus'],'k--',label='mean+stderr')
plt.plot(res_dt['e'], res_dt['minus'], 'k+',label='mean-stderr')
plt.plot(res_dt['e'],[max(res_dt['minus'].values)]*len(res_dt['e']),'r',label = 'reference')

plt.legend(loc=4)
ax.set_xlabel('Max Depth')
ax.set_ylabel('Mean Val AUC')
plt.title('X-validated AUC_DT by C')
plt.show()
```

```python
def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step
        
```

```python
# BernoulliNB

k = 8
alpha = [i for i in frange(0.1, 1.0, 0.1)]
auc_nb = xValNB(dataset_train, 'Rating', k, alpha)

xval_dict_nb = {'e':[], 'mean':[], 'stdErr':[],'plus':[],'minus':[]}
for i in alpha:
    xval_dict_nb['e'].append(i)
    xval_dict_nb['mean'].append(np.array(auc_nb[i]).mean())
    xval_dict_nb['stdErr'].append(sem(np.array(auc_nb[i])))
    xval_dict_nb['plus'].append(np.array(auc_nb[i]).mean()+sem(np.array(auc_nb[i])))
    xval_dict_nb['minus'].append(np.array(auc_nb[i]).mean()-sem(np.array(auc_nb[i])))
res_nb = pd.DataFrame(xval_dict_nb)

fig=plt.figure()
ax=fig.add_subplot(111)

plt.plot(res_nb['e'], res_nb['mean'],label='mean AUC')
plt.plot(res_nb['e'], res_nb['plus'],'k--',label='mean+stderr')
plt.plot(res_nb['e'], res_nb['minus'], 'k+',label='mean-stderr')
plt.plot(res_nb['e'],[max(res_nb['minus'].values)]*len(res_nb['e']),'r',label = 'reference')

plt.legend(loc=4)
ax.set_xlabel('Alpha')
ax.set_ylabel('Mean Val AUC')
plt.title('X-validated AUC_BernoulliNB by C')
plt.show()
```

```python
# LR
from sklearn.linear_model import LogisticRegression

k = 8
cs = [10**i for i in range(-7,5)]

auc_lr = xValLR(X_train, Y_train, k, cs)

xval_dict_lr = {'e':[], 'mu':[], 'sig':[], 'norm':[]}
for i in np.arange(-7,5):
    for norm in [1, 2]:
        xval_dict_lr['e'].append(i)
        xval_dict_lr['norm'].append(norm)
        xval_dict_lr['mu'].append(np.array(auc_lr[(10**i, norm)]).mean())
        xval_dict_lr['sig'].append(np.sqrt(np.array(auc_lr[(10**i, norm)]).var()))

xvals_lr = pd.DataFrame(xval_dict_lr)
xvals_lr['low'] = xvals_lr['mu']-xvals_lr['sig']/np.sqrt(k)
xvals_lr['up'] = xvals_lr['mu']+xvals_lr['sig']/np.sqrt(k)

norm = 1
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)], xvals_lr['mu'][(xvals_lr['norm']==norm)], 'r')
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)], xvals_lr['low'][(xvals_lr['norm']==norm)], 'r--', label='L1')
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)], xvals_lr['up'][(xvals_lr['norm']==norm)], 'r+')
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)],[max(xvals_lr['low'][xvals_lr['norm']==norm])]*len(xvals_lr[xvals_lr['norm']==norm]),'r',label = 'reference L1')

norm = 2
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)], xvals_lr['mu'][(xvals_lr['norm']==norm)], 'b')
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)], xvals_lr['low'][(xvals_lr['norm']==norm)], 'b--', label='L2')
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)], xvals_lr['up'][(xvals_lr['norm']==norm)], 'b+')
#plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)],[max(xvals_lr['low'][(xvals_lr['norm']==norm)])]*len([(xvals_lr['norm']==norm)]['e']),'b',label = 'reference L2')
plt.plot(xvals_lr['e'][(xvals_lr['norm']==norm)],[max(xvals_lr['low'][xvals_lr['norm']==norm])]*len(xvals_lr[xvals_lr['norm']==norm]),'b',label = 'reference L2')

plt.title('X-Validated AUC_LR by Regularization Weight')
plt.xlabel('Regularization weight C (logC)')
plt.ylabel('X-Val AUC')
plt.legend(loc=4)
plt.show()
```
